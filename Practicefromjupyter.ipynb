```python
Category_Data = df['Category'].astype(str)

Category_Codes = Category_Data.astype('category').cat.codes
plt.figure(figsize=(10, 5))
plt.hist(
    Category_Codes,
    bins=Category_Codes.nunique(),
    color='gray',
    edgecolor='red'
)
plt.xticks(rotation=45)   # Rotate 45 degrees
plt.title('Category Histogram (Encoded)')
plt.xlabel('Encoded Category')
plt.ylabel('Count')
plt.show()

```

```python
# now checking B2B data using the piechart 
import matplotlib.pyplot as plt

# Count how many times each value appears in the B2B column
b2b_counts = df['B2B'].value_counts()

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    b2b_counts,
    labels=b2b_counts.index,     # B2B values (like Yes/No)
    autopct='%1.1f%%',           # show percentages
    startangle=90,               # rotate starting angle
    colors=['#66b3ff', '#ff9999'], # nice contrasting colors
    explode=[0.05 if i==0 else 0 for i in range(len(b2b_counts))]  # pop out first slice
)

plt.title('B2B Distribution')
plt.axis('equal')  # make it a circle
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count occurrences in B2B column
b2b_counts = df['B2B'].value_counts()

plt.figure(figsize=(8, 8))

# Pie chart
wedges, texts, autotexts = plt.pie(
    b2b_counts,
    autopct='%1.1f%%',            # percentages on slices
    startangle=90,
    colors=['gray', 'purple'],
    wedgeprops={'edgecolor': 'black'},  # outline for clarity
    pctdistance=0.7                 # move percentage closer to center
)

# Add legend outside
plt.legend(
    wedges,
    b2b_counts.index,
    title="B2B",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1)  # place it to the right
)

plt.title('B2B Distribution')
plt.axis('equal')  # make it circular
plt.tight_layout()  # avoid cutting off
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Scatter plot
scatter_example = sns.scatterplot(
    x='Size',
    y='Category',
    data=df,
    hue='Category',        # optional: color by category
    palette='Set2',
    s=100                 # size of points
)

# Labels and title
scatter_example.set_xlabel('Size')
scatter_example.set_ylabel('Category')
scatter_example.set_title('Scatter plot: Size vs Category')

plt.xticks(rotation=45)    # rotate x-axis labels if needed
plt.tight_layout()
plt.show()
```

```python
# Get top 10 states by count
top_states = df['ship-state'].value_counts().nlargest(10).index

# Filter dataframe for only top 10 states
df_top10 = df[df['ship-state'].isin(top_states)]

plt.figure(figsize=(12, 6))

ax = sns.countplot(
    data=df_top10,
    x='ship-state',
    palette='Set2',
    order=top_states  # ensures bars are in descending order
)

# Add labels and title
ax.set_xlabel('ship-state', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.set_title('Top 10 Ship-States by City Count', fontsize=14)

# Rotate x labels for readability
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
